```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Örnek geçmiş ETH fiyatları (USD) - son 12 ay
eth_prices = [2000, 1800, 1900, 2500, 3000, 2800, 2600, 2200, 2300, 2700, 2900, 3100]

# Parametreler
monthly_investment = 500  # her ay yatırılacak miktar (USD)
months = len(eth_prices)

# Simülasyon
total_invested = 0
total_eth = 0
avg_price_list = []

for i in range(months):
    price = eth_prices[i]
    eth_bought = monthly_investment / price
    total_eth += eth_bought
    total_invested += monthly_investment
    avg_price = total_invested / total_eth
    avg_price_list.append(avg_price)
    print(f"Ay {i+1}: {price} USD, Alınan ETH: {eth_bought:.4f}, Toplam ETH: {total_eth:.4f}, Ortalama Fiyat: {avg_price:.2f} USD")

# Sonuçlar
print("\n=== Özet ===")
print(f"Toplam Yatırım: {total_invested} USD")
print(f"Toplam ETH Birikimi: {total_eth:.4f} ETH")
print(f"Ortalama Alım Fiyatı: {total_invested / total_eth:.2f} USD")

# Grafik
plt.figure(figsize=(10,5))
plt.plot(range(1, months+1), eth_prices, label='ETH Fiyatı')
plt.plot(range(1, months+1), avg_price_list, label='Ortalama Alım Fiyatı (DCA)')
plt.xlabel('Ay')
plt.ylabel('USD')
plt.title('ETH DCA Simülasyonu')
plt.legend()
plt.grid(True)
plt.show()
```
